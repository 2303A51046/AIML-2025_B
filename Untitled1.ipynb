```python
people = ['ann','bradon','chen','david','emily','farook',
          'gagan','hamish','imran','julio','katherine','liiy']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
weight = [55,35,68,77,70,60,72,69,18,65,82,48]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='red',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='green',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
